```python
import re
def extract_salaries(salary_str):
    salary_range = re.findall(r'\$(\d+)K', salary_str)
    if salary_range:
        lower_salary = int(salary_range[0]) * 1000
        upper_salary = int(salary_range[1]) * 1000 if len(salary_range) > 1 else None
        return pd.Series([lower_salary, upper_salary])
    return pd.Series([None, None])

dsjobs_df[['Lower Estimate','Upper Estimate']] = dsjobs_df['Salary Estimate'].apply(extract_salaries)
dsjobs_df
```

```python
dsjobs_df['Average Salary Estimate'] = (dsjobs_df['Lower Estimate'] + dsjobs_df['Upper Estimate'])/2
```

```python
def split_multicolumn(col_series):
    result_df = col_series.dropna().to_frame()
    
    options=[]
    
    temp_df=pd.DataFrame()
    
    for idx, value in col_series[col_series.notna()].items():
        for option in value.split(';'):
            if option not in temp_df.columns:
                options.append(option)
                temp_df[option] = False
            temp_df.at[idx,option]= True
            
    result =  pd.concat([result_df,temp_df],axis=1)
    return result_df[options]
```
